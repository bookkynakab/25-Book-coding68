```python
currency = input("ระบุสกุลเงิน (USD หรือ JPY): ")
amount = float(input("ระบุจำนวนเงิน: "))

if currency == 'USD':
    thb = amount * 32.57
    print("จำนวนเงินในหน่วยบาทคือ", thb)
elif currency == 'JPY':
    thb = amount * 0.29
    print("จำนวนเงินในหน่วยบาทคือ", thb
          )
else:
    print("ข้อมูลไม่ถูกต้อง กรุณาระบุเฉพาะ USD หรือ JPY")
```
